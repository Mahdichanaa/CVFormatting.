```python
import json

def input_list(prompt):
    print(prompt)
    items = []
    while True:
        item = input("Enter an item (or press enter to finish): ")
        if item == "":
            break
        items.append(item)
    return items

def input_dict(prompt, value_is_list=False):
    print(prompt)
    dictionary = {}
    while True:
        key = input("Enter a key (or press enter to finish): ")
        if key == "":
            break
        if value_is_list:
            value = input_list(f"Enter values for '{key}', separated by enters:")
        else:
            value = input(f"Enter value for '{key}': ")
        dictionary[key] = value
    return dictionary

def input_job():
    job = {
        "nom": input("Enter job name: "),
        "startDate": input("Enter start date: "),
        "endDate": input("Enter end date: "),
        "position": input("Enter position: "),
        "sujet": input("Enter sujet: "),
        "role": input("Enter role: "),
        "durée": input("Enter durée: "),
        "contexte": input("Enter contexte: "),
        "description": input("Enter description: "),
        "méthodologie": input("Enter méthodologie: "),
        "équipe": input("Enter équipe: "),
        "missions": input_list("Entrer les missions : "),
        "environnement": input_dict("Enter environnement : ", True)
    }
    return job 
```

```python
import datetime
def create_json():
    data = {
        "intro": {
            "position": input("Enter your position: "),
            "personName": input("Enter your name: "),
            "skills": input_list("Enter your skills (list of strings): "),
            "ans": input("Enter ans de competence (e.g., 4): "),
            "competence": input_list("Enter your competences (list of strings): "),
            "tech_skills": input_dict("Enter your tech skills (dictionary of lists): ", True),
            "formation": input_dict("Enter your formation (dictionary of strings): "),
            "langue": input_dict("Enter languages (dictionary of strings): ")
        },
        "jobs": []
    }

    while True:
        add_job = input("Do you want to add a job? (yes/no): ").lower()
        if add_job == "yes":
            job = {
                "nom": input("Enter job name: "),
                "startDate": input("Enter start date: "),
                "endDate": input("Enter end date: "),
                "position": input("Enter position: "),
                "sujet": input("Enter sujet: "),
                "role": input("Enter role: "),
                "durée": input("Enter durée: "),
                "contexte": input("Enter contexte: "),
                "description": input("Enter description: "),
                "méthodologie": input("Enter méthodologie: "),
                "équipe": input("Enter équipe: "),
                "missions": input_list("Enter missions (list of strings): "),
                "environnement": input_dict("Enter environnement (dictionary of lists): ", True)
            }
            data["jobs"].append(job)
        elif add_job == "no":
            break

    json_data= json.dumps(data, indent=4)
    filename = f"data_{datetime.datetime.now().strftime('%Y%m%d_%H%M%S')}.json"
    with open(filename,'w',encoding="utf-8") as file:
        file.write(json_data)
    return filename
```
